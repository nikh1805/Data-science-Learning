```python
import numpy as np
import matplotlib.pyplot as plt
# You can avoid using plt.show(). Automatically shows graphs
%matplotlib inline 
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
x, y
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
a, b
```

## Scatter plot

```python
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Scatter plot')
# plt.show()  # Not required as we have defined use inline matplotlib
```

## Plot

```python
y = x*x
```

```python
plt.plot(x,y,'r*--', linewidth=2, markersize=12)
```

## Subplots

```python
plt.subplot(2,2,1) # 2 rows, 2 columns, 1 postion
plt.plot(x,y,'r--')
plt.subplot(2,2,2) # 2 rows, 2 columns, 2 postion
plt.plot(x,y,'g*-')
plt.subplot(2,2,3) # 2 rows, 2 columns, 3 postion
plt.plot(x,y,'bo')
plt.subplot(2,2,4) # 2 rows, 2 columns, 3 postion
plt.plot(x,y,'y*--')
```

```python
x1 = np.arange(1,11)
y1 = 3 * x + 5
plt.plot(x1,y1)
```

### Sine wave form

```python
x = np.arange(0, 4 * np.pi, 0.1)
y = np.sin(x)
plt.title('Sine wave form')
plt.plot(x,y, 'g')
```

## Bar plots

```python
x = [2,8,10]
y = [11,16,9]
x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
```

## Histograms
#### https://en.wikipedia.org/wiki/Histogram

#### density/count on y axis. 

```python
a = np.array([22,87,5,43,56,73, 55, 54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title('Histogram')
plt.show()
```

### Box plot

#### To know percentile
#### lines -> 0,25,50,75,100, circles -> outliers

```python
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.boxplot(data, vert=True, patch_artist=True)
plt.show()
```

### Pie chart

```python
labels = 'python', 'c++', 'java', 'ruby'
sizes = [300, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1,0,0,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
